```python
import numpy.random as npr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Exponential Distribution
exp = npr.exponential(scale=0.05, size=10000)
plt.hist(exp, bins=100)
plt.show()
```

```python
# Half-Normal
halfnorm = npr.normal(loc=0, scale=10, size=40000)
halfnorm = halfnorm * (halfnorm > 0)
halfnorm = halfnorm[halfnorm != 0]
plt.hist(halfnorm, bins=100)
plt.show()
```

```python
# Gamma
gam = npr.gamma(shape=1, scale=1, size=10000)
plt.hist(gam, bins=100)
plt.show()
```
