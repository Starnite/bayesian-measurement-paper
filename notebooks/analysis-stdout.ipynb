```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context('poster')
sns.set_style('white')
```

```python
n_reps = pd.read_csv('sim_results/n_reps.txt', sep=": ", engine='python', header=None)
n_reps[1] = n_reps[1].str.strip(',').astype('int')
n_reps = n_reps[1]
n_reps.head()
```

```python
frac_accurate = pd.read_csv('sim_results/frac_correct.txt', sep=": ", engine="python", header=None)
frac_accurate = frac_accurate[1]
frac_accurate.head()
```

```python
data = pd.DataFrame()
data['n_reps'] = n_reps
data['frac_accurate'] = frac_accurate
sns.boxplot(x='n_reps', y="frac_accurate", data=data)
plt.xlabel('Num. Replicate Measurements')
plt.ylabel('Fraction Actual in HPD')
plt.savefig('../figures/accuracy.pdf', bbox_inches='tight', )
```
